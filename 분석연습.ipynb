```python
import pandas as pd

df = pd.read_csv("study_vs_score.csv")
print(df.head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.regplot(x='study_time', y='score', data=df)
plt.title("공부 시간과 점수 간의 관계")
plt.xlabel("study_time")
plt.ylabel("score")
plt.grid(True)
plt.show()

print("피어쓴 상관계수 : ", df.corr(method="pearson"))
print("스피어만 상관계수 : ", df.corr(method="spearman"))
print("켄달 상관계수 : ", df.corr(method="kendall"))
```

```python
import pandas as pd

df = pd.read_csv("salary_vs_satisfaction.csv")
print(df.head())
```

```python
import matplotlib.pyplot as plt

plt.scatter(df['salary'], df['satisfaction'])
plt.xlabel('salary')
plt.ylabel('satisfaction')
plt.title('Salary vs Satisfaction')
plt.show()
```

```python
import statsmodels.formula.api as smf

model = smf.ols('satisfaction ~ salary', data=df).fit()
print(model.summary())
```

```python
import pandas as pd

df = pd.read_csv("price_vs_purchase_balanced.csv")
print(df.head())
```

```python
import matplotlib.pyplot as plt

plt.scatter(df['price'], df['purchased'])
plt.xlabel('price')
plt.ylabel('purchased')
plt.title('Price vs Purchased')
plt.show()
```

```python
import statsmodels.api as sm

x = sm.add_constant(df['price'])
y = df['purchased']

model = sm.Logit(y, x).fit()
print(model.summary())
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("time_series_visitors.csv", index_col='date', parse_dates=True)

df['visitors'].plot(figsize=(10, 4), title="time_series_visitors")
plt.xlabel("date")
plt.ylabel("visitors")
plt.show()
```

```python
from statsmodels.tsa.stattools import adfuller

result = adfuller(df['visitors'])
print(result[0])
print(result[1])
```

```python
df_diff = df['visitors'].diff().dropna()
df_diff.plot(title="1차 차분 후 방문자 수 추이")
plt.show()
```

```python
result = adfuller(df_diff)
print(result[1])
```

```python
from statsmodels.tsa.arima.model import ARIMA

model = ARIMA(df['visitors'], order=(1,1,1))
model_fit = model.fit()

forecast = model_fit.forecast(steps=10)
print(forecast)
```
